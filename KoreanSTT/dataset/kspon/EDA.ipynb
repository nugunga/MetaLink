```python
def s(sentence):
    words = sentence[0].split()
    result = []
    for word in words:
        tmp = ''    
        for t in word:
            if not tmp:
                tmp += t
            elif tmp[-1]!= t:
                tmp += t
        result.append(tmp)
    return ' '.join(result)
```

```python
import datetime as dt
import os

now = dt.datetime.now()
formattedDate = now.strftime("%y%m%d_%H시%M분")
path = "D:/code/kospeech-latest/outputs"
result1 = "test_211210_03시27분(2021-12-09_03-14-09_epoch18).txt"
result2 = "test_211211_04시56분(2021-12-10_09-00-20_epoch30).txt"
    
with open(os.path.join(path, result1), 'r') as f:
    sentences1 = f.readlines()
with open(os.path.join(path, result2), 'r') as f:
    sentences2 = f.readlines()

with open(os.path.join(path, f"비교_{formattedDate}.txt"), 'a') as f:
    f.write(f"[{result1}]" + '  vs  ' + f"[{result2}]")
    f.write("\n\n")
    cnt = 0
    for (x, y) in zip(sentences1, sentences2):
        if x.split('\t')[-1] != y.split('\t')[-1]:
            f.write(x.split('\t')[0])
            f.write("\n")
            f.write("{}".format(x.split('\t')[-1]))
            f.write("{}".format(y.split('\t')[-1]))
            cnt +=1
    f.write("\n\n")
    f.write(f"총 다른 문장 {cnt}개")
```

```python
import shutil
import os
from random import sample
import json
import re
import pandas as pd


path_train = "D:/code/train wav"
path_meta = "D:/code/metadata"

train = os.listdir(path_train)
meta = os.listdir(path_meta)
labels_dest = "D:/code/kospeech-latest/dataset/kspon"

def rule(x):
    # 괄호
    a = re.compile(r'\([^)]*\)')
    # 문장 부호
    b = re.compile('[^가-힣0-9 ]')
    x = re.sub(pattern=a, repl='', string= x)
    x = re.sub(pattern=b, repl='', string= x)
    return x

def load_label(filepath):
    char2id = dict()
    id2char = dict()

    ch_labels = pd.read_csv(filepath, encoding="utf-8")

    id_list = ch_labels["id"]
    char_list = ch_labels["char"]
    freq_list = ch_labels["freq"]

    for (id_, char, freq) in zip(id_list, char_list, freq_list):
        char2id[char] = id_
        id2char[id_] = char
    return char2id, id2char

char2id, id2char = load_label(os.path.join(labels_dest, "cssiri_character_vocabs.csv"))

def sentence_to_target(sentence, char2id):
    target = str()

    for ch in sentence:
        try:
            target += (str(char2id[ch]) + ' ')
        # 사전에 없는 경우 넘어가라 -> 그냥 묵음처리나 마찬가지.
        except KeyError:
            continue

    return target[:-1]

for n in range(1,2):
    randints = sample(range(0,300000),1)
    dst = f'D:/code/sample_{n}'
    if not os.path.isdir(dst):
        os.mkdir(dst)
    if not os.path.isdir(os.path.join(dst, "audio")):
        os.mkdir(os.path.join(dst, "audio"))
    for num in randints:
        shutil.copy(os.path.join(path_train, train[num]), os.path.join(dst, "audio"))
    with open(path_meta + "/" + meta[num], 'r', encoding='UTF-8') as f:

        json_data = json.load(f)
        reading = json.dumps(json_data["transcription"]["ReadingLabelText"], ensure_ascii = False)
        reading = rule(reading)
        char_id_transcript = sentence_to_target(reading, char2id)
        with open(os.path.join(dst,"transcripts.txt"), "a") as f:
            # f.write('{0}\t{1}\n'.format(train[num], reading))
            f.write(f'{train[num]}\t{reading}\t{char_id_transcript}\n')
```
